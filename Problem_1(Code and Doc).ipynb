##Strategy
**STEP 2:**


Given a Truth Table for some Binanry operation on N bits...

We find the **SUM OF PRODUCT (SOP)** and **PRODUCT OF SUM (POS)** representation

**STEP 3:**
check following condition

**IF** (No. of Minterms in SOP <= No. of Maxterms in POS):
   We use SOP  

**ElSE**: We use POS   


**EXAMPLE 1**

image_2024-11-14_163735561.png

For the above tabel:

**SOP:-  F = ( !A . B . C) + (A . !B . C) + (A . B . !C)**

**POS:-  F = ( A + B + C) . (A + B + !C) . (A + !B + C) . (!A + B +C) . (!A +!B +C)**


CHECK CONDITION: STEP3

We see No. MINTERMS < No. MAXTERMS

Hence We work with Sop





**NOTE**:
**The Condition in STEP 3 can also be checked by counting the no. of 1's and 0's in the OUTPUT COLUMN F**

**IF** (No. of 1's in F column <= No. of 0's in F column)):
   We use SOP  

**ElSE**: We use POS

##Strategy
**STEP 5:**
**Create Ancilla Qubits and a Output Qubit:**

We need additnial qubits to store intermediate results of each AND/OR gate

For a "N" qubit input state and "K" no. of MAX/MINTERMS in the Boolean expression we would need an additional (K+2) qubits and a final Output qubit


We need K ancillas to store the values of Max/Minterms and 2 other qubits to evaluate the AND and OR operations iteratively

##Strategy
**STEP 6**:
**Create a Quantum Circuit for a Given SOP/POS expression:**

We use the "Quantum AND" , "Quantum OR" and "Quantum NOT" gates as seen in STEP 1 to implement a SOP or POS expression.

```python
def q_AND(qc,term,n):

  for i in range(len(term)):
    if term[i]=='0':
      qc.x(inputs[i])

  qc.mcx(inputs,ancill_K[n])

  for i in range(len(term)):
    if term[i]=='0':
      qc.x(inputs[i])

  qc.barrier()

  return
```

```python
def q_OR(qc):
  qc.x(ancill_K)
  qc.x(ancillas[0])
  qc.mcx(ancill_K,ancillas[0])
  qc.x(ancill_K)
  qc.barrier()
  return
```

```python
# For Truth Table in Example 1 with a SOP expression:- ( !A . B . C) + (A . !B . C) + (A . B . !C)
n=3
k=3
inputs = QuantumRegister(n , "inp")
ancillas = QuantumRegister(1, "ancil")
ancill_K = QuantumRegister(k, "Min/Max")
output = ClassicalRegister(n+1)
qc = QuantumCircuit(inputs,ancillas,ancill_K,output)
qc.reset(inputs)

# for input string 101
qc.x([inputs[0],inputs[2]])
qc.barrier(label="initialization to 101")
qc.draw("mpl")

```

```python
## Inplementation of SOP ( !A . B . C) + (A . !B . C) + (A . B . !C)

q_AND(qc,"011", 0) #the binary "011" represents the first minterm
q_AND(qc,"101", 1)
q_AND(qc,"110", 2)
q_OR(qc)

for i in range(n):
  qc.measure(inputs[i], output[i])
qc.measure(ancillas[0],output[n])

qc.draw("mpl")
```

```python
simulator_aer = AerSimulator()
qc_aer = transpile(qc,backend=simulator_aer)
# qc_aer.draw("mpl",idle_wires=False)
```

```python
result = simulator_aer.run(qc_aer,shots=10000).result()
counts = result.get_counts()
plot_distribution(counts)
```

**We see for input string 101 i.e Ψ= |101⟩  we get and output of 1 (rightmost bit) with 100% probability**

```python
#################### TO RUN CIRCUIT AGAIN FOR SAME TRUTH TABLE RUN ALL CELLS FROM HERE ON (Ctrl+F10) ###################################
n=N
k=bitstr.count('1')
inputs = QuantumRegister(n , "inp")
ancillas = QuantumRegister(2, "ancil")
ancill_K = QuantumRegister(k, "Mterms")
output = ClassicalRegister(n+1)
qc = QuantumCircuit(inputs,ancillas,ancill_K,output)

SOP =[]

for i in range(2**N):
  if bitstr[i] == '1':
    SOP.append(format(i, '0{}b'.format(N)))

inpstr='0'
while (len(inpstr)!=N):
  inpstr = input("Enter your desired input string: ")  ### we take Example 101
  if len(inpstr)!=N:print("enter valid input string")
qc.draw("mpl")
```

```python
def qc_OR(qc, first , second, store, nots):
  if nots[0]== '1':
    qc.x(first)
  if nots[1] == '1' :
    qc.x(second)
  qc.reset(store)
  qc.x([first,second,store])
  qc.ccx(first ,second,store)
  qc.x([first,second])
  if nots[0]== '1':
    qc.x(first)
  if nots[1] == '1' :
    qc.x(second)
  qc.barrier()
  return
```

```python
def qc_AND(qc, first , second, store, nots):
  if nots[0]=='0':
    qc.x(first)
  if nots[1] == '0':
    qc.x(second)
  qc.reset(store)
  qc.ccx(first, second,store)

  if nots[0]== '0':
    qc.x(first)
  if nots[1] == '0':
    qc.x(second)
  qc.barrier()

  return
```

**CIRCUIT GENERATION**

NOTE: The Circuit Beyond state preparation always remains the same for a given Truth Table.

**Qiskit doesn't allow arbitrary state prepartion without regenrating the rest of the circuit hence in this implementation the entire circuit is regerated if the user changes the input string. This won't actually be the case in practice as only the input state will be changed while keeping the rest of the circuit intact (for a given truth table)**

```python
for i in range(k):
  term = SOP[i]
  l =0
  for j in range(N-1):
    if (j == 0):
      qc_AND(qc,inputs[j], inputs[j+1], ancillas[l],[term[j],term[j+1]])
    elif j>0:
      qc_AND(qc,ancillas[l], inputs[j+1], ancillas[(l+1)%2],[1,term[j+1]])
      l = (l+1)%2
  qc.swap(ancillas[l],ancill_K[i])

l=0
for m in range(k-1):
      if (m == 0):
        qc_OR(qc,ancill_K[m], ancill_K[m+1], ancillas[l],[0,0])
      elif(m>0):
        qc_OR(qc,ancillas[l], ancill_K[m+1], ancillas[(l+1)%2],[0,0])
        l = (l+1)%2


for i in range(N):
  qc.measure(inputs[i], output[i])
qc.measure(ancillas[l],output[N])
qc.draw("mpl")
```

```python
for i in range(N):
  qc.measure(inputs[i], output[i])

simulator_aer1 = AerSimulator()
qc_aer1 = transpile(qc,backend=simulator_aer1)
result = simulator_aer1.run(qc_aer1,shots=10).result()
counts = result.get_counts()
plot_distribution(counts)


########## to run again for diffrent input or for a diffrent truth table altogether go back to previous cells at the beginning ##################
```
